 13. Perform following aggregation task on below table <br>
        a. Get the average age group by Region <br>
        b. Get the sum of Age and Validation group by Hair color and then reset the index <br>
        c. Get the man and the median of Age group by Region

```python
t = pd.DataFrame([[25,'Black','Southwest',1],[30,'Black','Northwest',1],[45,'Red','Northeast',1],[35,'Blond','Southwest',0],[33,'Brown','Southwest',0],[35,'Brown','Northeast',0],[25,'Black','Southwest',0],[52,'Brown','Southeast',0],[21,'Blond','Southwest',1],[24,'Blond','Northeast',0],[48,'Black','Southwest',1],[22,'Black','Southwest',0],[27,'Brown','Northeast',0],[28,'Black','Southeast',1],[21,'Blond','Southeast',1],[51,'Black','Northwest',0],[23,'Brown','Southeast',1]],columns=['Age','Hair Color',
'U.S. Region','Validation'])
```

```python
t
```

```python
t.groupby('U.S. Region')['Age'].mean()
```

14. Split the below table into half and perform merge (inner, left and right) on Region. One example for each type of merge (inner, left and right)

```python
t1 = t[0:9]
t1
```

```python
t2 = t[9:19]
t2
```

```python
t2
```

```python
t1.merge(t2, how='inner',on='U.S. Region')
```

```python
t1.merge(t2, how='left',on='U.S. Region')
```

```python
t1.merge(t2, how='right',on='U.S. Region')
```
